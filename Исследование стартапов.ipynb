## 2. Инжиниринг признаков
### 2.1 Задача 1 
##### Группы по срокам финансирования. Разделите все компании на три группы: Единичное финансирование — был всего один раунд финансирования. Срок финансирования до года — между первым и последним раундом финансирования прошло не более года. Срок финансирования более года. Визуализируйте соотношение этих групп, создав два графика: По количеству компаний: покажите, какой процент от общего числа компаний относится к каждой из трёх групп. По объёму инвестиций: отобразите, какую долю от общего объёма привлечённых средств получила каждая группа.

````python
df['funding_interval'] = (df['last_funding_at'] - df['first_funding_at']).dt.days
df.loc[df['funding_rounds'] == 1, 'funding_group'] = 'Единичное финансирование'
df.loc[(df['funding_rounds'] > 1) & (df['funding_interval'] <= 365), 'funding_group'] = 'Срок до года'
df.loc[(df['funding_rounds'] > 1) & (df['funding_interval'] > 365), 'funding_group'] = 'Срок более года'

display("Распределение компаний по группам:")
display(df['funding_group'].value_counts())
````

````python
counts = df['funding_group'].value_counts()
labels = counts.index
sizes_counts = counts.values

total_by_group = df.groupby('funding_group')['funding_total_usd'].sum()
sizes_total = total_by_group.values

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.pie(sizes_counts, labels=labels, autopct='%1.1f%%', startangle=90,
        colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Доля компаний по группам', fontsize=14)

plt.subplot(1, 2, 2)
plt.pie(sizes_total, labels=labels, autopct='%1.1f%%', startangle=90,
        colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Доля общего объёма инвестиций по группам', fontsize=14)

plt.tight_layout()
plt.show()
````

````python
market_counts = df['market'].value_counts()

df['market_group'] = df['market']

df.loc[df['market'].isin(market_counts[market_counts <= 35].index), 'market_group'] = 'niche'
df.loc[df['market'].isin(market_counts[(market_counts > 35) & (market_counts <= 120)].index),'market_group'] = 'mid'

df['is_outlier'] = False

for segment in df['market_group'].dropna().unique():
    segment_mask = df['market_group'] == segment
    segment_data = df.loc[segment_mask, 'funding_total_usd']
    
    Q1 = segment_data.quantile(0.25)
    Q3 = segment_data.quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    df.loc[
        segment_mask & (
            (df['funding_total_usd'] < lower) |
            (df['funding_total_usd'] > upper)
        ),
        'is_outlier'
    ] = True

outlier_stats = df.groupby('market_group').agg(total=('is_outlier', 'count'),
                                           outliers=('is_outlier', 'sum')
)

outlier_stats['share'] = outlier_stats['outliers'] / outlier_stats['total']
outlier_stats = outlier_stats.sort_values('share', ascending=False)

display('Доля выбросов по сегментам:')
display(outlier_stats)

display('Топ сегментов по доле выбросов:')
display(outlier_stats.head(10))
````

На основе получившейся таблицы постройте графики, которые отразят: динамику типичного размера средств, которые стартапы получали в рамках одного раунда финансирования;  динамику общего количества раундов за каждый год, то есть насколько активно происходили инвестиции на рынке (чем больше раундов, тем выше активность). Когда будете строить графики в этом задании и следующих, используйте данные только по тем компаниям, которые остались в датасете после предыдущих фильтраций. На основе полученных данных ответьте на вопросы: В каком году типичный размер средств, собранных в рамках одного раунда, был максимальным? Какая тенденция наблюдалась в 2014 году по количеству раундов и средств, выделяемых в рамках каждого раунда?

````python
df_clean = df.copy()

df_clean = df_clean[(df_clean['funding_rounds'] > 0) & (df_clean['mid_funding_at'].notna()) & (df_clean['funding_total_usd'].notna())].copy()

df_clean['mid_funding_at'] = pd.to_datetime(df_clean['mid_funding_at'], errors='coerce')

df_clean = df_clean[df_clean['mid_funding_at'].notna()].copy()

df_clean['avg_round_size'] = df_clean['funding_total_usd'] / df_clean['funding_rounds']

df_clean['year'] = df_clean['mid_funding_at'].dt.year

yearly_stats = (df_clean.groupby('year').agg(median_round_size=('avg_round_size', 'median'),
                                             total_rounds=('funding_rounds', 'sum')
).reset_index().sort_values('year'))

display('Статистика по годам:')
display(yearly_stats)
````

````python
plt.figure(figsize=(10, 5))
plt.plot(
    yearly_stats['year'],
    yearly_stats['median_round_size'],
    marker='o',
    linestyle='-'
)
plt.title('Динамика медианного размера одного раунда финансирования по годам')
plt.xlabel('Год')
plt.ylabel('Медианный размер раунда, USD')
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 5))
plt.bar(yearly_stats['year'], yearly_stats['total_rounds'], alpha=0.7)
plt.title('Динамика общего количества раундов финансирования по годам')
plt.xlabel('Год')
plt.ylabel('Количество раундов')
plt.grid(axis='y')
plt.show()
````

````python
max_row = yearly_stats.loc[yearly_stats['median_round_size'].idxmax()]
display(
    f"Максимальный медианный размер одного раунда был в {int(max_row['year'])} году: "
    f"{max_row['median_round_size']:,.0f} USD"
)


if 2014 in yearly_stats['year'].values:
    row_2014 = yearly_stats[yearly_stats['year'] == 2014].iloc[0]

    prev_years = yearly_stats[yearly_stats['year'] < 2014]

    if not prev_years.empty:
        avg_prev_rounds = prev_years['total_rounds'].mean()
        avg_prev_size = prev_years['median_round_size'].mean()

        display('Показатели за 2014 год:')
        display(f"Медианный размер одного раунда: {row_2014['median_round_size']:,.0f} USD")
        display(f"Общее количество раундов: {row_2014['total_rounds']:,.0f}")

        display('Средние значения за предыдущие годы:')
        display(f"Средний медианный размер раунда: {avg_prev_size:,.0f} USD")
        display(f"Среднее количество раундов: {avg_prev_rounds:,.0f}")

        if row_2014['median_round_size'] < avg_prev_size and row_2014['total_rounds'] > avg_prev_rounds:
            display('Вывод: в 2014 году типичный размер раунда был ниже среднего, а активность рынка по числу раундов — выше средней.')
        elif row_2014['median_round_size'] < avg_prev_size and row_2014['total_rounds'] <= avg_prev_rounds:
            display('Вывод: в 2014 году типичный размер раунда был ниже среднего, и активность рынка по числу раундов не выросла.')
        elif row_2014['median_round_size'] >= avg_prev_size and row_2014['total_rounds'] > avg_prev_rounds:
            display('Вывод: в 2014 году и типичный размер раунда, и активность рынка были на высоком уровне.')
        else:
            display('Вывод: в 2014 году показатели были близки к среднему уровню предыдущих лет.')
else:
    display('Данных за 2014 год нет.')
````
